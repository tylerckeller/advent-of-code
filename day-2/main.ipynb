```python
from collections import defaultdict


def valid_games(r, g, b, data):
    game_sum_result = 0
    game_num = 0
    pulls = []
    for line in data:
        game_num = int(line.split(':')[0].split(' ')[1])
        pulls = line.split(':')[1].split(';')
        game_pulls = []
        for pull in pulls:
            pull_vals = pull.split(',')
            pull_dict = {}
            for val in pull_vals:
                val = val.strip()
                pull_dict[val.split(' ')[1]] = int(val.split(' ')[0])
            game_pulls.append(pull_dict)
        valid = True
        print(game_num, game_pulls)
        i = 0
        while valid:
            pull_dict = game_pulls[i]
            for k, v in pull_dict.items():
                if k == 'red':
                    if r - v < 0:
                        valid = False
                        break
                if k == 'green':
                    if g - v < 0:
                        valid = False
                        break
                if k == 'blue':
                    if b - v < 0:
                        valid = False
                        break
            if i == len(game_pulls) - 1:
                break
            i += 1
        if valid:
            game_sum_result += game_num
    return game_sum_result
```

```python
def minimum_valid_set(data):
    game_num = 0
    pulls = []
    game_num_dict = {}
    for line in data:
        game_num = int(line.split(':')[0].split(' ')[1])
        pulls = line.split(':')[1].split(';')
        game_pulls = []
        for pull in pulls:
            pull_vals = pull.split(',')
            pull_dict = {}
            for val in pull_vals:
                val = val.strip()
                pull_dict[val.split(' ')[1]] = int(val.split(' ')[0])
            game_pulls.append(pull_dict)
        game_num_dict[game_num] = game_pulls
    min_valid_set = {}
    for k, v in game_num_dict.items():
        min_r, min_g, min_b = 0, 0, 0
        game_num = k
        for pull in v:
            for k, v in pull.items():
                if k == 'red':
                    min_r = max(min_r, v)
                if k == 'green':
                    min_g = max(min_g, v)
                if k == 'blue':
                    min_b = max(min_b, v)
        min_valid_set[game_num] = {'red': min_r, 'green': min_g, 'blue': min_b}
    return min_valid_set
```

```python
def compute_min_valid_set_power_sum(min_valid_set):
    min_valid_set_power_sum = 0
    for single_min_valid_set in min_valid_set.values():
        min_valid_set_power_sum += (single_min_valid_set['red'] * single_min_valid_set['green'] * single_min_valid_set['blue'])
    return min_valid_set_power_sum
```
