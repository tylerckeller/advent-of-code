```python
def load_data(datafile):
    data = []
    with open(datafile) as f:
        data =  f.read().splitlines()
    seeds = [int(x) for x in data[0].split(':')[1].strip().split(' ')]
    seed_to_soil = {}
    soil_to_fertilizer = {}
    fertilizer_to_water = {}
    water_to_light = {}
    light_to_temperature = {}
    temperature_to_humidity = {}
    humidity_to_location = {}
    all_mappings = [seed_to_soil, soil_to_fertilizer, fertilizer_to_water, water_to_light, light_to_temperature, temperature_to_humidity, humidity_to_location]
    map_i = 0
    i = 1
    while i < len(data):
        if data[i] == '':
            # hit a map
            # increase to get past map title
            i += 2
            cur_map = all_mappings[map_i]
            while i < len(data) and data[i] != '':
                # ingest data
                line = data[i].split(' ')
                destination_range_start = int(line[0])
                source_range_start = int(line[1])
                range_len = int(line[2])
                cur_map[destination_range_start] = (source_range_start, range_len)
                i += 1
            # add mapping
            all_mappings[map_i] = dict(sorted(cur_map.items()))
        map_i += 1
    return seeds, all_mappings

load_data('input')
```
