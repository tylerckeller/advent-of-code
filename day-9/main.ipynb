```python
sequences = []
for line in data:
    sequences.append([int(x) for x in line.split()])
```

```python
# find the differences between consecutive numbers and put them in a list
# continue until the differences are all the 0
# then, use that list to find the next number in the original sequence
# return the sum of the next numbers in each sequence
from collections import defaultdict
sum = 0
all_differences = {}
sequence = sequences[0]
sequence_diff = []
inner_diff = sequence
print(sequence)
while not all([x == 0 for x in inner_diff]):
    inner_diff = []
    for i in range(len(sequence) - 1):
        inner_diff.append(sequence[i+1] - sequence[i])
    sequence_diff.append(inner_diff)
all_differences[tuple(sequence)] = sequence_diff

all_differences
```
