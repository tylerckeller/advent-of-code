```python
def parse_input(datafile):
    data = []
    parsed_data = {}
    with open(datafile, 'r') as f:
        data = f.readlines()
    for line in data:
        line = line.strip()
        card_id = int(line.split(':')[0].strip().split(' ')[-1])
        winning_cards = line.split(':')[1].split('|')[0].strip().split(' ')
        player_cards = line.split(':')[1].split('|')[1].strip().split(' ')
        winning_cards = [int(i) for i in winning_cards if i != '']
        player_cards = [int(i) for i in player_cards if i != '']
        set_winning_cards = set(winning_cards)
        set_player_cards = set(player_cards)
        matching_cards = set_winning_cards.intersection(set_player_cards)
        parsed_data[card_id] = {'matching_cards': matching_cards}
    return parsed_data

def calculate_score(parsed_data):
    score = 0
    for card_id in parsed_data.keys():
        prior_score = score
        score += 2**(len(parsed_data[card_id]['matching_cards']) - 1) if len(parsed_data[card_id]['matching_cards']) > 0 else 0
        print(card_id, parsed_data[card_id]['matching_cards'], score-prior_score)
    return score

calculate_score(parse_input('input'))
```

```python
from collections import defaultdict


def calculate_num_cards(parsed_data):
    '''
        Iterate through parsed_data which represents the original cards.
        For each card, check the number of matching cards. The number of matched cards is equal to the number of won cards below the current card id.
        Ex. card_id = 1, matched_cards = 2, then won_cards = [2, 3]
        Iterate through the won_cards and check the number of matched cards for each won card. The number of matched cards is equal to the number of won cards below their current card id.
        Continue until there are no more won cards.
        Then, move to the next original card.
        Return the total number of cards won.
    '''

    # setup dictionary to store the number of won cards for each card
    card_to_num_won_cards = {}
    for card_id, card_data in parsed_data.items():
        card_to_num_won_cards[card_id] = len(card_data['matching_cards'])

    # setup frequency dictionary to store the number of won cards for each card
    card_freq = defaultdict(lambda: 1)
    for card_id in parsed_data.keys():
        card_freq[card_id]

    for card_id in parsed_data.keys():
        for i in range(card_id+1, card_id+card_to_num_won_cards[card_id]+1):
            card_freq[i] += 1*card_freq[card_id]

    return sum(card_freq.values())

calculate_num_cards(parse_input('input'))
```
